```python
# Indexing -> start from 0

Array indexing = pointing to a LEGO block by its seat number (always starting from 0)

What is array indexing?

Array indexing means accessing a value from an array using its position number.

1. In Python (and NumPy), indexing starts from 0.

2. Negative indexing (from the end)

   You can count from the end using negative numbers.

3. Indexing in a 2D array (rows & columns). 

4. Use [] to access elements.


```

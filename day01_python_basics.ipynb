```python
#Formatted Strings
first = "Diya"
last = "Mathew"
full = f"{first} {last}"
print(full)
```

```python
full = F"{len(first)} {2+2}"
print(full)
```

```python
full = f"{len(first)} {first} {last}"
print(full)
```

```python
# In a formatted string (f-string), spaces are not automatic.

# f is a prefix that tells Python:

# “This string contains expressions inside {}. Evaluate them.”

# It is part of Python’s string syntax, not a function.

full = f"{first }{last}"
print(full)
```

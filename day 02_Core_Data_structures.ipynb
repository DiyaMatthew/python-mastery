```python
# List

courses = ["History","Computer Science","Biology","Physics"]
print(courses)
```

```python
print(courses[1])
```

```python
# Negative Indexes

print(courses[-1])
```

```python
print(courses[-2])
```

```python
print(courses[0])
```

```python
# Modifying List -> List methods
# add one item at the end


courses.append("Art")
print(courses)
```

```python
# to insert an item to a particular location -> insert()

courses.insert(1,"AI")
print(courses)
```

```python
# to add another list -> extend()

new_courses = ["Data Science","Robotics"]
courses.extend(new_courses)
print(courses)
```

```python
print(courses[6])
```

```python
# Looping Values

for course in courses:
    print(course)
```

```python
for index,course in enumerate(courses):
    print(index,course)
```

```python
for index,course in enumerate(courses, start=1):
    print(index,course)
```

```python
# Remove an item from the list -> remove()

courses.remove("Art")
print(courses)

# Remove all items ->clear()
```

```python
print(courses[7])
```

```python
# Remove item from the end of the list and return that item -> pop()

popped=courses.pop()
print(popped)
print(courses)
```

```python
# sorting list ->sort()

courses.sort()
print(courses)
```

```python
courses.reverse()
print(courses)
```

```python
courses.sort(reverse=True)
print(courses)
```

```python
print(courses)
```

```python
courses.sort()
print(courses)
```

```python
print(courses.sort(reverse=True))
```

```python
print(courses)
```

```python
# To find the indexes of an item in the list -> index()

index_course = courses.index("Data Science")
print(courses)
print(index_course)
```

```python
# Convert list to strings

course_str = " - ".join(courses)
print(course_str)
```

```python
course_str = " ".join(courses)
print(course_str)
```

```python
# convert back to list

new_list = course_str.split(" ")
print(new_list)

```

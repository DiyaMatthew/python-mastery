```python
# Conditional Statements

temperature = 35
if temperature>30:
    print("it's warm")
    print("Drink Plenty of water")
elif temperature>20:
    print("Dream weather")
else:
    print("it's cold")
print("Done")    
```

```python
# Ternary Operator
# Without Ternary Operator

age = 22
if age>18:
    print("Eligible")
else:
    print("Not eligible")
```

```python
# With Ternary Operator
# A ternary operator is a short, one-line way to write an ifâ€“else condition
# Itâ€™s also called a conditional expression

age = 22
message = "Eligible" if age>=18  else "Not Eligible" # Correct syntax: value_if_true if condition else value_if_false
print(message)


```

```python
# Logical Operators
# and
# or
# not

high_income = True
good_credit = True
student = False
if (high_income or good_credit)and not student:
    print("Eligible")
else:
    print("Not Eligible")    
```

```python
# Short Circuit Evaluation
# In python logical operators are short circuit

if (high_income or good_credit)and not student:
    print("Eligible")
else:
    print("Not Eligible")    
```

```python
# Chaining Comparison operators
# age should be between  18 and 65

if 18<= age <65:
    print("Eligible")
```

# While Loop

while loop vs for loop
Use "for loop" when the number of iterations is known, and "while loop" when looping depends on a condition.

Example 

ðŸ½ï¸ Eating biscuits

for loop â†’ You have 5 biscuits

while loop â†’ You eat until you are full

```python
command = ""
while command.lower()!="quit":
    command = input()
    print("Echo", command)
```

```python
# Infinite Loop
while True:
    command = input(">")
    print("Echo", command)
    if command.lower()=="quit":
        break
    
```

# Functions 

A function is a block of reusable code that does a specific task.

Instead of writing the same code again and again, you write it once and reuse it.

```python
# Functions

def greet():
    print("Hello")
    print("Welcome on Board")


  
```

```python
greet()
```

```python
# Arguments
Arguments vs Parameters

# ðŸ‘‰ Parameters are the names/variables written inside a function definition.
  They are like empty boxes waiting for values.


# ðŸ‘‰ Arguments are the actual values you pass to the function when calling it.
  They are the real data that goes into the boxes.
```

```python
# defining a funaction


def greet(first_name, last_name):
    print(f"Hello {first_name} {last_name}")
    print("Welcome Aboard")
    
```

```python
# Calling a Function


greet("Diya","Mathew") 
```

```python
# Types of Functions

1. Perform a task
   eg. print()
2. Return a value
   eg. round()
```

```python
# 1. perform a task functions


def greet(first_name, last_name):
    print(f"Hi {first_name} {last_name}")
```

```python
greet("Diya", "Mathew")
```

```python
# 2. Return a Value


def get_greeting(name):
    return f"Hi {name}"
```

```python
get_greeting("Diya Mathew")
```

* Keyword arguments are arguments passed using parameter_name = value.

Instead of giving values by position, you explicitly tell Python which value belongs to which parameter.

```python
# Keyword Arguments

def increment(number, by):
    return number + by
```

```python
print(increment(2, by=1))
```

# Default Arguments

Default arguments are parameters that already have a value.
If you donâ€™t pass an argument, Python uses the default value.

```python
# Default Arguments

def increment(number, by=1):   # All optional parameters come after required parameters
    return number + by
    
```

```python
increment(2)
```

```python
increment(2,4)
```

*args â†’ Multiple positional arguments

ðŸ‘‰ What is *args?

Allows a function to accept any number of positional arguments

Python stores them as a tuple

ðŸ§  Remember

args is just a name (you can use any name)

* is what matters

```python
# *args

def multiply(*numbers):
    total = 1
    for number in numbers:
        print(number)
        total *= number
    return total    
```

```python
print(multiply(1,2,3,4,5))
```

```python
ðŸ”¹ **kwargs â†’ Multiple keyword arguments

ðŸ‘‰ What is **kwargs?

Allows a function to accept any number of keyword arguments

Python stores them as a dictionary
```

```python
def show_details(**kwargs):
    print(kwargs)

show_details(name="Diya", qualification="Advanced MSc", role="AI Engineer")
```
